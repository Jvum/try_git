```python
# Importação de componentes
import pandas as pd
import numpy as np

# Sklearn Libraries - importação das bibliotecas
from sklearn.preprocessing import Imputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Importação do Plot
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Leitura do arquivo a ser analisado
dataset = pd.read_csv('test_data.csv')
```

```python
# Mostrar que o DataFreme está no Pandas
type(dataset)
```

```python
# Exibir os valores do arquivo importado
dataset
```

```python
# Definição de variáveis dependentes e indepentes do modelo

x = dataset.iloc[:,0].values
y = dataset.iloc[:,-1].values

# Valor negativo indica última, penúltima... coluna

x = x.reshape((-1,1))
```

```python
# Mostra os valores de 'y'
y
```

```python
# Treinamento do Regressor - criação da variável de regressão
regressor = LinearRegression()
regressor.fit(x,y)
# variável independente é 'x'
```

```python
x
```

```python
# Chamar função de criação do gráfico - 'pyplot' foi nomeado de 'plt'

plt.figure(figsize=(12,9))
plt.scatter(x,y,color='red', s=10)
plt.plot(x, regressor.predict(x), color = 'blue')
plt.title('Avaliação - Test Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
