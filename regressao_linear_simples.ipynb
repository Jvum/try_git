### Importa todas as bibliotecas necessarias

```python
import pandas as pd
import numpy as np


# Sklearn libraries
from sklearn.preprocessing import Imputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression


from matplotlib import pyplot as plt
## Comando para que o grafico plotado apareça logo abaixo do código que o produz
%matplotlib inline
```

### Lê o arquivo de dados

```python
dataset = pd.read_csv('test_data.csv')
```

```python
type(dataset)
```

### Exibe o arquivo de dados na tela

```python
dataset
```

### Separa as variáveis independentes e dependentes do problema

```python
X = dataset.iloc[:,0].values
y = dataset.iloc[:,-1].values
X = X.reshape((-1,1))
```

```python
y
```

### Treina o regressor

```python
regressor = LinearRegression()
regressor.fit(X, y)
```

### Visualiza o resultado do treinamento

```python
X
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(X, y, color = 'orange', s=60)
plt.plot(X, regressor.predict(X), color = 'purple')
plt.title('Gráfio X vs Y')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```
